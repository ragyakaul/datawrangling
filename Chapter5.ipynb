```python
# UPDATING THE CODE TO FIX DOUBLE-LINED ERRORS (WITH PREVIOUS LINE)

pdf_txt = 'en-final-table9.txt'
openfile = open(pdf_txt, "r")

double_lined_countries = [
    'Bolivia (Plurinational \n',
    'Democratic People\xe2\x80\x99s \n',
    'Democratic Republic \n',
    'Lao People\xe2\x80\x99s \n',
    'Micronesia (Fedrated \n', 
    'Saint Vincent and \n',
    'The former Yugoslav \n',
    'United Republic \n',
    'Venezuela (Bolivarian \n',
]

def turn_on_off(line, status, start, prev_line, end='\n'): # The first three, line, status, and start, are required arguments—that is, they must be provided because they have no default values. The last one, end, has a default value of a newline character, as that appears to be a pattern with our file. We can override the default value by passing a different value when we call our function.
    """
            This function checks to see if a line starts/ends with a certain
            value. If the line starts/ends with that value, the status is
            set to on/off (True/False).
    """
    if line.startswith(start): # Replaces what we are searching for on the starting line with the start variable.
        status = True
    elif status: # Replaces the end text we used with the end variable.
        if line == end and prev_line != 'and areas':
            status = False # then we can turn data collection off
    return status # Returns the status based on the same logic (end means False, start means True).

def clean(line):
    """
    Cleans line breaks, spaces, and special characters from our line
    """
    line = line.strip('\n').strip() # Strips \n off of the line and reassigns the output to line so now line holds thecleaned version
    line = line.replace('\xe2\x80\x93', '-') # Replaces special character encodings
    line = line.replace('\xe2]x80\x99', '\'') # Returns our newly cleaned string
    
    return line


countries = []
totals = []
country_line = total_line = False
previous_line = ''


for line in openfile:
    if country_line: # If it’s a country line, this line adds the country to the country list.
        print('%r' % line) # Adds a print '%r' statement to output the Python representation
        if previous_line in double_lined_countries:
            line = ' '.join([clean(previous_line), clean(line)]) # If the previous_line is in the double_lined_countries list, this line joins the previous_line with the current line and assigns the combined lines to the line variable
            countries.append(line)  
        elif line not in double_lined_countries: # If the line is not in the double_lined_countries list, then the following line adds it to the countries list.
            countries.append(clean(line))
            
    elif total_line: # This line collects totals, the same as we did with countries.
        if len(line.replace('\n', '').strip()) > 0: # We know from experience the PDF uses newlines as blank lines. On this line, the code replaces newlines with nothing and strips whitespace to clean it. Then this code tests whether the string still has a length greater than zero. If so, we know there is data and we can add it to our data (totals) list.
            totals.append(clean(line))
    
    country_line = turn_on_off(line, country_line, previous_line, 'and areas')
    total_line = turn_on_off(line, total_line, previous_line, 'total')
    
    previous_line = line
    
    
# We are going to combine the totals and countries by “zipping” the two datasets. 
# The zip function takes an item from each list and pairs them together until all items are paired. 
# We can then convert the zipped list to a dictionary by passing it to the dict function.
import pprint
data = dict(zip(countries, totals)) # Zips the countries and totals lists together by calling zip(countries, totals). This turns them into a tuple. We then change the tuple into a dictionary, or dict (for easier reading), by passing it to the dict function.
pprint.pprint(data) 
```
